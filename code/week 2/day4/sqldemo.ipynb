```python
import pymssql
from sqlalchemy import create_engine
import pandas as pd
```

```python
#get environment variables
server = "DESKTOP"
database = "assignment_1"

#establish the connection using windows authenticate
conn = pymssql.connect(
    server=server,
    user='',
    password='',
    database=database
)

#creates the cursor object
cursor = conn.cursor()

#executing a query
cursor.execute('select * from dbo.categories')

#fetch all the results
rows = cursor.fetchall()

#print the resuts
for row in rows:
    print(rows)

#put data into dataframe
columns = [col[0] for col in cursor.description]
df = pd.DataFrame(rows, columns=columns)

#close cursor and connection
cursor.close()
conn.close()
```

```python
#create dataframe
data = {
    'column1': [1,2,3],
    'column2': ['A','B','C']
}
df = pd.DataFrame(data)

#sql server connection
server = "DESKTOP"
database = "assignment_1"

#create the engine
engine = create_engine(f'mssql+pymssql://@{server}/{database}')

#write the dataframe to sql server table
    # if_exists options
    # - replace- replace the existing table by that name
    # - append- adds this table onto the existing table by that name
    #       -can create problems or a lot of null values- if they dont have fields in common
    # - fail- if it exists, don't replace or append, just reject it
df.to_sql('new_table3', engine, if_exists='replace', index=False)

#close connection to engine
engine.dispose
```
