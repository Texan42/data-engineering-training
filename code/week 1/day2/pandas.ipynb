Series:
- ID labeled array (column of data)
- can hold and data type
- adds an index to each row
- flexible
- powerful built in method


```python
s= pd.Series([1,2,3,4,5])
print(s)
```

```python
s= pd.Series([1,2,3,4,5], index= ['a','b', 'c', 'd', 'e'])
print(s)
```

```python
s1 = pd.Series([1,2,3], index=['a','b','c'])
s2 = pd.Series([4,5,6], index=['b','c','d'])

print(s1)
print(s2)
```

```python
print(s1+s2)
```

Different formats in python using pandas and dataframes
- 'dot notation' df.total_rooms
- 'bracket notation' df['total_rooms'] <---- best practice

why bracket notation:
- python/pandas has reserved words (i.e. boxplot) wont throw an error if a column has a reserved word
- .notation can lead python to think you're referring to a function NOT a column
