```python
import pandas as pd

df = pd.read_csv('imdb_top_1000.csv', encoding='ISO-8859-1')
df.head()
```

```python
df.isnull().mean() * 100
```

```python
df.dtypes
```

```python
df.columns.to_list()
```

```python
original_columns = ['Poster_Link',
 'Series_Title',
 'Released_Year',
 'Certificate',
 'Runtime',
 'Genre',
 'IMDB_Rating',
 'Overview',
 'Meta_score',
 'Director',
 'Star1',
 'Star2',
 'Star3',
 'Star4',
 'No_of_Votes',
 'Gross']

new_columns = [col.lower() for col in original_columns]
df.columns = new_columns
```

```python
df.columns.to_list()
```

#1

```python
columns_to_drop = ['poster_link']
df.drop(columns=columns_to_drop, inplace=True)
```

```python
df.info()
```

```python
df['released_year'] = df['released_year'].str.extract('(\d+)').astype(float)
```

```python
df['released_year'] = pd.to_numeric(df['released_year'])
```

```python
df.dtypes
```

#2 & 3

```python

print('The oldest movie was released in:')
print(df['released_year'].min().astype('int64'))

print('The newest movie was released in:')
print(df['released_year'].max().astype('int64'))
```

#4

```python
df_sortby_rating = df.sort_values(by='imdb_rating', ascending=False)
df_sortby_rating.head(10)
```

#5

```python
df_sortby_genre  = df_sortby_rating.drop_duplicates(subset='genre', keep='first')
df_sortby_genre.head(10)
```
